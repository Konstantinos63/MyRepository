# https://data.cdc.gov/Behavioral-Risk-Factors/Behavioral-Risk-Factor-Surveillance-System-BRFSS-P/y4ft-s73h/about_data

# first importing the necessary libraries: pandas for handling data and requests for making HTTP requests.
# then load a CSV file called mycldata.csv into a pandas DataFrame (mycldata). 
# sending a GET request to a CDC API (https://data.cdc.gov/resource/y4ft-s73h.json) to retrieve data in JSON format, with a limit of 1 million records, and storing the response in the response variable.

```python
import pandas as pd
import requests
path='/Users/kostaskal'
mycldata=pd.read_csv(path+'/mycldata.csv')
url="https://data.cdc.gov/resource/y4ft-s73h.json?$limit=1000000"
response=requests.get(url)
response
```

```python
payload=response.json()
payload
dfapi=pd.DataFrame(payload)
dfapi
```

# In the below code, I first define two lists: states (a list of state names) and years (a list of years). I then filter the dfapi DataFrame to include only rows where the state is one of the listed states and the year is one of the listed years. After filtering, I make a copy of the DataFrame and convert the year column to integers. I create a new variable, postpolicy, which is 0 for the years 2005-2007 and 1 for 2009-2010. Finally, I create a dummy variable NY_dummy for New York (1 if locationabbr is 'NY', 0 otherwise) and a new interaction variable, NY_postpolicy_interaction, as the product of NY_dummy and postpolicy.

```python
states = ['New Jersey', 'New York', 'Florida', 'Pennsylvania', 'Massachusetts','Connecticut']
years = ['2005', '2006', '2007', '2009', '2010']
dfapi_filtered = dfapi[
    (dfapi["locationdesc"].isin(states)) &
    (dfapi["year"].isin(years))
]
dfapi_filtered.year.unique()
dfapi_filtered = dfapi_filtered.copy()
dfapi_filtered['year'] = dfapi_filtered['year'].astype(int)
dfapi_filtered.loc[:, 'postpolicy'] = dfapi_filtered['year'].apply(
    lambda x: 0 if x in [2005, 2006, 2007] else 1
)
dfapi_filtered['NY_dummy'] = dfapi_filtered['locationabbr'].apply(
    lambda x: 1 if x == 'NY' else 0
)
dfapi_filtered['NY_postpolicy_interaction'] = dfapi_filtered['NY_dummy'] * dfapi_filtered['postpolicy']
dfapi_filtered
```

```python
dfapi_filtered.year.unique()
```

# In this code, I drop two columns ('data_value_footnote_symbol' and 'data_value_footnote') from the dfapi_filtered DataFrame to create a new DataFrame called dfapi_cleaned. I then export the dfapi_cleaned DataFrame to a CSV file named dfapi_cleaned.csv, ensuring that the row indices are not included in the output. This results in a clean CSV file with the specified columns removed.

```python
dfapi_cleaned = dfapi_filtered.drop(columns=['data_value_footnote_symbol', 'data_value_footnote'])
dfapi_cleaned.to_csv("dfapi_cleaned.csv", index=False)
```

```python
dfapi_cleaned
```

# In this code, I create a dictionary statefips that maps FIPS codes to state abbreviations. I then use the map() function on the STATEFIP column of mycldata to replace the FIPS codes with the corresponding state abbreviations, storing the result in a new column called state. The updated DataFrame mycldata now contains the state column with the state abbreviations.

```python
statefips= {
    36: "NY",
    12: "FL",
    34: "NJ",
    9:  "CT",
    25: "MA",
    42: "PA"
}
mycldata["state"] = mycldata["STATEFIP"].map(statefips)
mycldata
```

# FRY-OVERLOAD. The lower the value of the ratio, we assume the higher the frequency of fries consumption for the individual.
# P.S: FRIESMO.y --> frequency of fries consumption during last month.

```python
mycldata['FRIESMO.y']
```

```python
mycldata['fryoverload'] = mycldata['BMI'] / (mycldata['FRIESMO.y'] + 1)
mycldata
mycldata.to_csv("mycldata.csv", index=False)
```

# In this code, I first print the data types of the columns in dfapi and mycldata to inspect their structure. I then clean the year column in dfapi by converting it to a string, extracting only the 4-digit years, and removing any rows with missing values. Next, I convert the locationabbr column to uppercase in both DataFrames and ensure that the YEAR column in mycldata is of integer type. Finally, I rename columns in dfapi to match those in mycldata, and perform an inner join to merge the two DataFrames on the YEAR and state columns, storing the result in merged_df.

```python
print(dfapi.dtypes)
print(mycldata.dtypes)
dfapi_cleaned["year"] = dfapi_cleaned["year"].astype(str).str.extract("(\d{4})")  # keep only 4-digit years
dfapi_cleaned = dfapi_cleaned[dfapi_cleaned["year"].notna()]  # remove rows with missing year
dfapi_cleaned["year"] = dfapi_cleaned["year"].astype(int)
dfapi_cleaned["locationabbr"] = dfapi_cleaned["locationabbr"].str.upper()
mycldata["YEAR"] = mycldata["YEAR"].astype(int)
mycldata["state"] = mycldata["state"].str.upper()

dfapi_renamed = dfapi_cleaned(columns={
    "year": "YEAR",
    "locationabbr": "state"
})

merged_df = pd.merge(mycldata, dfapi_renamed, on=["YEAR", "state"], how="inner")
merged_df
```
