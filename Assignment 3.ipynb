```python
# Importing my datasets. However, I only use the mycldata because I have numerical values there. I will plot the time series analysis for BMI for the years 2005 through 2011. My legends will be : NY,NJ,FL,MA,PA,CT
```

```python
import pandas as pd
dfapi=pd.read_csv("dfapi_cleaned.csv")
dfapi
dfbmi=pd.read_csv("mycldata.csv")
dfbmi
dfbmi.state.unique()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Filter for relevant states
states = ['NY', 'NJ', 'FL', 'PA', 'MA', 'CT']
filtered_dfbmi = dfbmi[dfbmi['state'].isin(states)]
# Group by state and year, then calculate average BMI
avg_bmi = filtered_dfbmi.groupby(['state', 'YEAR'])['BMI'].mean().reset_index()
```

```python
# Set up the side-by-side plots (line and bar)
fig, axs = plt.subplots(1, 2, figsize=(16, 6), sharey=True)
# Line plot of average BMI by year for each state
sns.lineplot(data=avg_bmi, x='YEAR', y='BMI', hue='state', marker='o', ax=axs[0])
axs[0].set_title('Average BMI Over Time by State (Line Plot)')
axs[0].set_xlabel('Year')
axs[0].set_ylabel('Average BMI')
axs[0].legend(title='State', bbox_to_anchor=(1.05, 1), loc='upper left')
```

# This code puts three graphs side by side to help us understand how BMI has changed over time in six states. The first graph is a line chart that shows the average BMI in each state from 2005 to 2011. The second one is a bar chart showing the same info but in a different style. The third graph is a boxplot, which helps us see how individual BMI values are spread out in each state  like whether most people have similar BMIs or if there are a lot of differences. Putting all three together makes it easier to spot trends and compare states.

```python
# Prepare the 3 subplots in one row
fig, axs = plt.subplots(1, 3, figsize=(24, 6), sharey=False)

# Line plot of average BMI by year for each state
sns.lineplot(data=avg_bmi, x='YEAR', y='BMI', hue='state', marker='o', ax=axs[0])
axs[0].set_title('Average BMI Over Time by State (Line Plot)')
axs[0].set_xlabel('Year')
axs[0].set_ylabel('Average BMI')
axs[0].legend(title='State', bbox_to_anchor=(1.05, 1), loc='upper left')

# Bar plot of average BMI by year for each state
sns.barplot(data=avg_bmi, x='YEAR', y='BMI', hue='state', ax=axs[1])
axs[1].set_title('Average BMI Over Time by State (Bar Plot)')
axs[1].set_xlabel('Year')
axs[1].set_ylabel('Average BMI')
axs[1].legend(title='State', bbox_to_anchor=(1.05, 1), loc='upper left')

# Boxplot for individual BMI distribution by state
sns.boxplot(data=filtered_dfbmi, x='state', y='BMI', ax=axs[2])
axs[2].set_title('BMI Distribution by State (20052011)')
axs[2].set_xlabel('State')
axs[2].set_ylabel('Individual BMI')

plt.tight_layout()
plt.show()
```
