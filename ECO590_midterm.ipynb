### Question 9. Restauraunts per Zipcode
1. Using the NYCOpenData API, get all restaurant inspections data: https://data.cityofnewyork.us/Health/DOHMH-New-York-City-Restaurant-Inspection-Results/43nn-pn8j/about_data
    - API Endpoint: https://data.cityofnewyork.us/resource/43nn-pn8j.json
    - Each row is a restauraunt inspection. It contains __duplicate restauraunts__.
    - The camis column is a unique identifier for each restauraunt.
2. How many unique restaurants are there in this data?
3. Create a dataframe that has the count of unique restauraunts per zipcode.

```python
import requests # pandas already imported from beginning #Q1
url='https://data.cityofnewyork.us/resource/43nn-pn8j.json'
response=requests.get(url) # sending client a response to retrieve data
response # code status is 200, good to go
payload=response.json() #json format of the response
payload
df1=pd.DataFrame(payload) # converting json format to a dataframe
df1
```

```python
df1=df1.drop_duplicates() # dropping duplicates and also filling NaN values with 0
dfnonan=df1.fillna(0)     # assigning df to a new object, whereby I filled NaN with 0
dfnonan.isna()           # i don't drop NaN because I'm loosing so many rows, I rather fill NaN with 0
```

```python
dfnonan['dba'].nunique() # Q2 measuring the n:number of unique restaurants, dba column gives me restaurant names
```

```python
df2=dfnonan.groupby('zipcode').nunique('dba') # Q3, I group the dfnonan (clean dataset) by zipcode
df3=df2.dba                                       #in the same line I meaasure the number of unique restaurants(dba)
df3=pd.DataFrame(df3)
df3
```
